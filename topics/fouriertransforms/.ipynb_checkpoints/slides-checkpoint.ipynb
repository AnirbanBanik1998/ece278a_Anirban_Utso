    <h2 style="font-family: Trebuchet MS; margin-bottom: -20px"><center>Derivation</center></h2>
    <p><center><i>Background</i></center></p>
    <ul>
        <li>A <b>complex</b> number is expressed in its rectangular form as <i>z = x + jy</i>.</li><br>
        <li>The <b>polar form</b> expresses a complex number in terms of its magnitude and phase in the complex plane as <i>z = rcos&theta; + jrsin&theta;</i>.</li><br>
        <li><b>Sinusoids</b> describe circular motion. They are important because they naturally occur everywhere and are therefore fundamental in physics. Any physical system that resonates or oscillates can be characterized by sinusoidal motion, so it is often useful to express data in its complex polar form.</li><br>
        <li><b>Euler's identity</b> compactly expresses the polar form of a complex number as an exponential. This makes solving differential equations much easier.</li>
    </ul>
</div>

<div style="font-family: Trebuchet MS">
    <h2 style="font-family: Trebuchet MS; margin-bottom: -20px"><center>Derivation</center></h2>
    <p><center><i>Background</i></center></p>
    <ul>
        <li>Euler's identity states that <i>e<sup>j&theta;</sup> = cos&theta; + jsin&theta;</i>.</li><br>
        <li>To prove Euler's identity, observe the Taylor Series Expansion of each term.</li>
    </ul>
    <center><img style="width:60%" src="euler.png"></center>
    <ul>
        <li>A <b>complex exponential</b> describes the polar form of a complex number using Euler's identity as <i>re<sup>j&theta;</sup> = rcos&theta; + rsin&theta;</i>. A complex exponential can be regarded as the tracing of a circle in the complex plane, where <i>r</i> is the radius of the circle and <i>&theta;</i> describes the current position of the trace.</li>
    </ul>
</div>

    <ul>

        <li>The <b>inner product</b> of two complex vectors <i>u</i> and <i>v</i> is defined as</li><br>

        <li>The inner product can be regarded as a mapping from two length N complex vectors to a complex scalar. Two vectors are <b>orthogonal</b> if their inner product is zero.</li><br>
        <li>In <b>continuous time</b>, samples are thought to be infinitely close to one another. This means an integral is needed to take the inner product of two continuous-time vectors.</li><br>
    </ul>
</div>

    <p><center><i>Geometric Signal Theory</i></center></p>
    <ul>
        <li>A <b>domain</b> describes the specific set of N-dim basis vectors used to represent a signal with N samples.</li><br>
        <li>In the <b>time domain</b>, there are N basis vectors, where each basis vector corresponds to an impulse at a certain moment in time.</li><br>

    <h2 style="font-family: Trebuchet MS; margin-bottom: -20px"><center>Derivation</center></h2>
    <p><center><i>Geometric Signal Theory</i></center></p>
    <ul style="font-size:18px">
        <li>The <b>Fourier Transform</b> is a change-of-base process from the time domain to the frequency domain. It computes the inner product of a given time signal with several sinusoidal basis vectors. These inner products are plotted on the x-axis as coefficients, shedding light on the frequency content of the signal. In practice, we write sinusoids as complex exponentials using Euler's identity. Therefore, these inner products yeild complex coefficients. When analyzing a signal in the frequency domain, this means we can plot the coefficients' magnitude, phase, real part, or imaginary part.</li><br>
        <li>The <b>Inverse Fourier Transform</b> reconstructs the original signal in time by taking the linear combination of the sinusoidal basis vectors in the frequency domain after scaling them by their inner product with the original signal. This reconstruction process works because sinusoids of different frequencies are orthogonal to each other.</li><br>
        <li>The Fourier Transform is often described as a way to approximate a time signal using a linear combination of sinusoids. This description comes from the freedom to choose the number of basis vectors in the frequency domain; if fewer basis vectors are used, the reconstruction is less accurate.</li>
    </ul>

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fs = 200
t = np.arange(0,1,1/fs)
freq1 = 8 # Hz
freq2 = 14 # Hz
f = np.cos(2*np.pi*t*freq1) + np.cos(2*np.pi*t*freq2)
plt.plot(t,f); plt.xlabel('Time (s)'); plt.ylabel('Amplitude')
plt.title(str(freq1)+' Hz cosine wave plus '+str(freq2)+' Hz cosine wave')
plt.show()
```

```python
N = len(t);
k = np.arange(0,fs,fs/N)
F = np.fft.fft(f,N)
plt.plot(k[:int(len(k)/4)],F[:int(len(F)/4)])
plt.xlabel('Freq (Hz)')
plt.ylabel('Amplitude')
plt.title('Fourier Transform of the signal')
plt.show()
```

<div style="font-family: Trebuchet MS">
    <h2 style="font-family: Trebuchet MS; margin-bottom: -20px"><center>Forms</center></h2>
    <p><center><i>Types of Fourier Transforms</i></center></p>
</div>

- Fourier transforms are integral transforms that takes in an arbitrary input function in the time domain and transforms it into the frequency domain.

- They can be continuous or discrete in terms of input time and output frequency and can be across finite or infinite time durations.

- Conventional Fourier transforms work with continuous time and frequency for an infinite duration, which necessitates continuous integrals.

$$
X(\omega) = \int_{-\infty}^{\infty} x(t)e^{j \omega t} dt, \omega \in (-\infty, +\infty)
$$ 
- $\omega$ is the output frequency and $x(t)$ is the continuous time input function

- The complex coefficient actually _simplifies_ the equation due to Euler's identity and not needing to find 2 real coefficients for each frequency.


- The input function $x(t)$ would look something like $x(t) = \frac{1}{2}a_0 + \sum_{k = 1}^\infty (a_k cos (2\pi k t) + b_k sin(2 \pi k t))$

<div style="font-family: Trebuchet MS">
    <h2 style="font-family: Trebuchet MS; margin-bottom: -20px"><center>Forms</center></h2>
    <p><center><i>Types of Fourier Transforms</i></center></p>
</div>

- For other inputs like images, we can not represent them in terms of continuous functions, so we have to convert the transform to work with discretized inputs. 

$$
X(k) = \sum_{n=0}^{N-1} x(n)e^{-j\omega_k n}, k = 0, 1, ..., N - 1
$$ 
- Each k represents a frequency bin determined by dividing the sampling frequency by the number of samples taken.

<div style="font-family: Trebuchet MS">
    <h2 style="font-family: Trebuchet MS"><center>Example</center></h2>
    <ul><li><p>The slides below illustrate calculating the DFT on a simple sine wave. [9]</p></li></ul>
</div>

```python
def fft(seq, samp_freq, samples):
    bins = np.zeros(samp_freq,dtype=np.complex_)
    for x in range(samples):
        for y in range(samp_freq):
            bins[x] += seq[y] * (np.cos(-2*np.pi*x*y/samples) + 1j*np.sin(-2*np.pi*x*y/samples))
    return bins

def mag(complex_seq):
    return np.sqrt(np.power(complex_seq.real, 2) + np.power(complex_seq.imag, 2))
```

```python
true_x = np.linspace(0, 2*np.pi, 100)
true_y = np.sin(true_x)
plt.plot(true_x, true_y, label="sin(t)")

x_values = np.linspace(0, 2*np.pi, 9)[:-1]
y_values = np.sin(x_values)
plt.scatter(x_values, y_values)
plt.show()
```

```python
fft_y_values = fft(y_values, 8, 8)

mag_y = mag(fft_y_values)

x_bins = np.linspace(0, 7, 8, dtype=int)
plt.scatter(x_bins, mag_y)
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude")
plt.title("DFT of sine wave")
plt.show()
```

```python
fft_y_values = fft(y_values, 8, 8)
fft_y_values = fft_y_values*2
mag_y = mag(fft_y_values)
mag_y = mag_y[:len(mag_y)//2]

x_bins = np.linspace(0, 7, 8, dtype=int)
x_bins = x_bins[:len(x_bins)//2]
plt.scatter(x_bins, mag_y)
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude")
plt.title("Actual DFT of sine wave")
plt.show()
```

<div style="font-family: Trebuchet MS">
    <h2 style="font-family: Trebuchet MS"><center>Things to note</center></h2>
</div>

- Potential pitfalls occur when the Nyquist sampling theorem is not followed.

- Aliasing can occur when you do not sample at at least double the frequency of the highest possible frequency found in the input function.

- This higher frequency component can then appear in the DFT result errantly, masking other true frequency components.

- Can apply anti-aliasing filters or raise the sampling frequency.

<div style="font-family: Trebuchet MS">
    <h2 style="font-family: Trebuchet MS"><center>Example 2</center></h2>
    <ul><li><p>The slides below illustrate interesting behaviors when performing FFTs on images. </p></li></ul>
    <ul><li><p>Fast Fourier Transforms (FFTs) are just faster algorithms for computing the discrete Fourier Transforms.</p></li></ul>
</div>

```python
img_1_open = Image.open('lena.jpg')
img_2_open = Image.open('baboon.jpg')

img_1 = np.asarray(img_1_open.convert("L"))
img_2 = np.asarray(img_2_open.convert("L"))

#FFT of images
fft_img_1 = np.fft.fft2(img_1)
fft_img_2 = np.fft.fft2(img_2)

#Switching the phase components
phase_angle1 = np.angle(fft_img_1)
phase_angle2 = np.angle(fft_img_2)

new_fft_img_1 = np.abs(fft_img_1) * np.exp(1j * phase_angle2)
new_fft_img_2 = np.abs(fft_img_2) * np.exp(1j * phase_angle1)

#Reconstructing images
ifft_img_1 = np.fft.ifft2(new_fft_img_1).real
ifft_img_2 = np.fft.ifft2(new_fft_img_2).real
```

```python
fig, axs = plt.subplots(1, 4)

axs[0].imshow(img_1, cmap="gray")
axs[0].set_title("img_1")

axs[1].imshow(img_2, cmap="gray")
axs[1].set_title("img_2")

axs[2].imshow(ifft_img_1, cmap="gray")
axs[2].set_title("ifft_img_1")

axs[3].imshow(ifft_img_2, cmap="gray")
axs[3].set_title("ifft_img_2")

for i in range(4):
    axs[i].axis('off')
    
plt.show()
```

<div style="font-family: Trebuchet MS">
    <h2 style="font-family: Trebuchet MS"><center>What's Happening?</center></h2>
</div>

- The predominant features of each image gets swapped after switching the phase response before running the inverse FFT.

- The phase contains information regarding how the different frequencies contribute to the structure of the image. 

- The magnitude just tells how much each frequency is actually contributing, which means the phase is more important for the overall shape/features of the image.

- Fourier transforms are also useful in solving partial differential equations systems.

- In essence, neural operators use neural networks to learn to solution operators for PDEs. So, given the initial or boundary conditions, the neural network can directly output the solution.

- This differs from conventional PDE solvers that need to discretize the space into a fine mesh, whereas the neural operators are mesh-independent.

- The neural operator will learn the continuous function rather than discretized vectors.

<p style="font-family: Trebuchet MS">Import</p>

```python
import torch
import matplotlib.pyplot as plt
import sys
from neuralop.models import TFNO
from neuralop import Trainer
from neuralop.datasets import load_darcy_flow_small
from neuralop.utils import count_model_params
from neuralop import LpLoss, H1Loss

device = 'cpu'
```

<p style="font-family: Trebuchet MS">Load the Darcy-flow dataset</p>

```python
train_loader, test_loaders, data_processor = load_darcy_flow_small(
        n_train=3000, batch_size=32, 
        test_resolutions=[16, 32], n_tests=[1200, 600],
        test_batch_sizes=[32, 32],
        positional_encoding=True
)
data_processor = data_processor.to(device)
```

<p style="font-family: Trebuchet MS">Create a tenorized Fourier Neural Operator</p>

```python
model = TFNO(n_modes=(32, 32), hidden_channels=32, projection_channels=64, factorization='tucker', rank=0.42)
model = model.to(device)

n_params = count_model_params(model)
print(f'\nOur model has {n_params} parameters.')
sys.stdout.flush()
```

<p style="font-family: Trebuchet MS">Create the optimizer</p>

```python
optimizer = torch.optim.Adam(model.parameters(), 
                                lr=8e-3, 
                                weight_decay=1e-4)
scheduler = torch.optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=30)
```

<p style="font-family: Trebuchet MS">Create the losses</p>

```python
l2loss = LpLoss(d=2, p=2)
h1loss = H1Loss(d=2)

train_loss = h1loss
eval_losses={'h1': h1loss, 'l2': l2loss}
```

```python
print('\n### MODEL ###\n', model)
print('\n### OPTIMIZER ###\n', optimizer)
print('\n### SCHEDULER ###\n', scheduler)
print('\n### LOSSES ###')
print(f'\n * Train: {train_loss}')
print(f'\n * Test: {eval_losses}')
sys.stdout.flush()
```

<p style="font-family: Trebuchet MS">Create the trainer</p>

```python
trainer = Trainer(model=model, n_epochs=20,
                  device=device,
                  data_processor=data_processor,
                  wandb_log=False,
                  log_test_interval=3,
                  use_distributed=False,
                  verbose=True)
```

<p style="font-family: Trebuchet MS">Train the model on the Darcy-flow dataset</p>

```python
trainer.train(train_loader=train_loader,
              test_loaders=test_loaders,
              optimizer=optimizer,
              scheduler=scheduler, 
              regularizer=False, 
              training_loss=train_loss,
              eval_losses=eval_losses)
```

<p style="font-family: Trebuchet MS">Plot the prediction, and compare with the ground-truth</p>

```python
test_samples = test_loaders[32].dataset

fig = plt.figure(figsize=(7, 7))
for index in range(3):
    data = test_samples[index]
    data = data_processor.preprocess(data, batched=False)
    # Input x
    x = data['x']
    # Ground-truth
    y = data['y']
    # Model prediction
    out = model(x.unsqueeze(0))

    ax = fig.add_subplot(3, 3, index*3 + 1)
    ax.imshow(x[0], cmap='gray')
    if index == 0: 
        ax.set_title('Input x')
    plt.xticks([], [])
    plt.yticks([], [])

    ax = fig.add_subplot(3, 3, index*3 + 2)
    ax.imshow(y.squeeze())
    if index == 0: 
        ax.set_title('Ground-truth y')
    plt.xticks([], [])
    plt.yticks([], [])

    ax = fig.add_subplot(3, 3, index*3 + 3)
    ax.imshow(out.squeeze().detach().numpy())
    if index == 0: 
        ax.set_title('Model prediction')
    plt.xticks([], [])
    plt.yticks([], [])

fig.suptitle('Inputs, ground-truth output and prediction.', y=0.98)
plt.tight_layout()
plt.show()
```
